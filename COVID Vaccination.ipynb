```python
import pandas as pd
data = pd.read_csv("/home/leong/Desktop/Covid Vaccination/vaccination-data.csv")
data.head()
```

Clean Data

```python
data.isnull().sum()
```

```python
#Fill NUMBER_VACCINES_TYPES_USED  with mcolumn mean
data['TOTAL_VACCINATIONS'].fillna(data['TOTAL_VACCINATIONS'].mean(), inplace=True)

#Fill FIRST_VACCINE_DATE  with mode
data['FIRST_VACCINE_DATE'].fillna(data['FIRST_VACCINE_DATE'].mode()[0], inplace=True)

#Fill PERSONS_BOOSTER_ADD_DOSE   with mcolumn mean
data['PERSONS_BOOSTER_ADD_DOSE'].fillna(data['PERSONS_BOOSTER_ADD_DOSE'].mean(), inplace=True)

#Fill TOTAL_VACCINATIONS_PER100 with mean
data['TOTAL_VACCINATIONS_PER100'].fillna(data['TOTAL_VACCINATIONS_PER100'].mean(), inplace = True)

#Fell NUMBER_VACCINES_TYPES_USED with mode
data['NUMBER_VACCINES_TYPES_USED'].fillna(data['NUMBER_VACCINES_TYPES_USED'].mode()[0], inplace = True)

#Fill PERSONS_BOOSTER_ADD_DOSE   with mcolumn mean
data['PERSONS_BOOSTER_ADD_DOSE_PER100'].fillna(data['PERSONS_BOOSTER_ADD_DOSE_PER100'].mean(), inplace=True)


```

Visualizations

```python
#General Data visualization
import matplotlib.pyplot as plt
data.hist(figsize=(12, 10), bins=20)
plt.show()
```

```python

import seaborn as sns
sns.boxplot(data=data[['TOTAL_VACCINATIONS', 'TOTAL_VACCINATIONS_PER100']])
plt.show()
```

```python
#Time series analysys
data['DATE_UPDATED'] = pd.to_datetime(data['DATE_UPDATED'])
data.set_index('DATE_UPDATED', inplace=True)
data['TOTAL_VACCINATIONS'].plot(figsize=(12, 6))
plt.title('Total Vaccinations Over Time')
plt.show()
```

```python
#Regional Analysis
region_counts = data['WHO_REGION'].value_counts()
region_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Regional Distribution of Data')
plt.xlabel('WHO Region')
plt.ylabel('Number of Records')
plt.show()
```

```python
#Correlation Analysis
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
#Personwise analysis
plt.figure(figsize=(300, 300))
vaccinated_status_counts = data['PERSONS_VACCINATED_1PLUS_DOSE'].value_counts()
vaccinated_status_counts.plot(kind='pie', autopct='%1.1f%%', figsize=(8, 8))
plt.title('Distribution of Vaccination Status')
plt.show()
```

Comparative aacross different countries/regions

```python

# Comparative analysis by country
plt.figure(figsize=(20, 20))
sns.barplot(x='COUNTRY', y='TOTAL_VACCINATIONS_PER100', data=data, ci=None)
plt.xticks(rotation=90)
plt.title('Total Vaccinations Per 100 People by Country')
plt.show()

```

```python
# Comparative analysis by region
plt.figure(figsize=(10, 6))
sns.barplot(x='WHO_REGION', y='TOTAL_VACCINATIONS_PER100', data=data, ci=None)
plt.title('Total Vaccinations Per 100 People by WHO Region')
plt.show()
```

I dentifying outliers or success stories by examining countries or regions with unusually high or low vaccination rates compared to their peers.

```python
# Identify outliers by country
outliers_by_country = data[data['TOTAL_VACCINATIONS_PER100'] > data['TOTAL_VACCINATIONS_PER100'].mean() + 2 * data['TOTAL_VACCINATIONS_PER100'].std()]
outliers_by_country
```

```python
# Identify outliers by region
outliers_by_region = data[data.groupby('WHO_REGION')['TOTAL_VACCINATIONS_PER100'].transform(lambda x: (x > x.mean() + 2 * x.std()))]
outliers_by_region
```
